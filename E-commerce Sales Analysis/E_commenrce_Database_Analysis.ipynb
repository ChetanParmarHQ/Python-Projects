#Converting date columns

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df[['Order Date', 'Ship Date']].info()
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Day of Week'] = df['Order Date'].dt.dayofweek
```

```python
df[['Order Date','Order Month', 'Order Year', 'Order Day of Week']].head()
```

# Monthly Sales Analysis

```python
sales_by_month = df.groupby('Order Month')['Sales'].sum().reset_index()

sales_by_month
```

```python
fig = px.line(sales_by_month,
              x = 'Order Month',
              y = 'Sales',
              title = 'Monthly Sales Analysis')
fig.show()
```

# Sales by Category

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()

sales_by_category
```

```python
from IPython.utils import text
fig = px.pie(sales_by_category,
             values = 'Sales',
             names = 'Category',
             hole = 0.5)

fig.update_traces(textposition = 'inside', textinfo = 'percent+label')
fig.update_layout(title_text = 'Sales Analysis by Category', title_font=dict(size=24))

fig.show()
```

#Sales Analysis by Sub Category

```python
sales_by_subcategory = df.groupby('Sub-Category')['Sales'].sum().reset_index().sort_values(by = 'Sales', ascending = False)

sales_by_subcategory
```

```python
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory,
             x = 'Sub-Category',
             y = 'Sales',
             title = 'Sales Analysis by Sub-Category')
fig.show()
```

# Monthly Profit Analysis

```python
profit_by_month = df.groupby('Order Month')['Profit'].sum().reset_index()

profit_by_month
```

```python
fig = px.bar(profit_by_month,
             x = 'Order Month',
             y = 'Profit',
             title = 'Monthly Profit Analysis')
fig.show()
```

# Profit by Category

```python
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()

profit_by_category
```

```python
fig = px.pie(profit_by_category,
             values = 'Profit',
             names = 'Category',
             hole = 0.5)

fig.update_traces(textposition = 'inside', textinfo = 'percent+label')
fig.update_layout(title_text = 'Profit by Category', title_font = dict(size=24))

fig.show()
```

#Profit by Sub Category

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by = 'Profit', ascending = False)

profit_by_subcategory
```

```python
fig = px.bar(profit_by_subcategory,
             x = 'Sub-Category',
             y = 'Profit',
             title = 'Profit Analysis by Sub Category')
fig.show()
```

# Sales and Profit - Customer Segment

```python
df['Segment'].value_counts()
```

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()

sales_profit_by_segment
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(x = sales_profit_by_segment['Segment'],
                     y = sales_profit_by_segment['Sales'],
                     name = 'Sales'))
fig.add_trace(go.Bar(x = sales_profit_by_segment['Segment'],
                     y = sales_profit_by_segment['Profit'],
                     name = 'Profit'))
fig.update_layout(title = 'Sales and Profit Analysis by Segment',
                  xaxis_title = 'Custoemr Segment', yaxis_title = 'Amount')
fig.show()
```

# Sales to Profit Ratio

```python
# For Segment

sales_profit_by_segment['sales_profit_ratio'] = sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']

print(sales_profit_by_segment[['Segment', 'sales_profit_ratio']])
```

```python
# For Category

sales_profit_by_category = df.groupby('Category').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()

sales_profit_by_category['sales_profit_ratio'] = sales_profit_by_category['Sales']/sales_profit_by_category['Profit']

print(sales_profit_by_category[['Category', 'sales_profit_ratio']])
```

```python
# For Sub Category

sales_profit_by_subcategory = df.groupby('Sub-Category').agg({'Sales':'sum', 'Profit': 'sum'}).reset_index()

sales_profit_by_subcategory['sales_profit_ratio'] = sales_profit_by_subcategory['Sales']/sales_profit_by_subcategory['Profit']

print(sales_profit_by_subcategory[['Sub-Category', 'sales_profit_ratio']])
```
